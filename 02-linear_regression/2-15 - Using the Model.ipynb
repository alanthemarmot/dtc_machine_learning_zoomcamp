```python
# Linear Regression formulae, write in code -> formulae in notes

#takes the list of features as the input
def linear_regression(xi):

    #get the number of features to test
    n = len(xi)

    # set the prediction value, initialised to the base weight
    pred = w0

    #loop through the features and get the outcome of the weight multiplied by the feature
    for j in range(n):
        pred = pred + w[j] * xi[j]

    #return the end prediction
    return pred
```

### 2.9 - Root Mean Square Error(RMSE)

Looking to quantify how good/bad the model is. Root Mean Squared Error (RMSE) is a way to evaluate regression models. It measures the error associated with the model being evaluated. This numerical figure then can be used to compare the models, enabling us to choose the one that gives the best predictions.

Formulae: $ \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2} $

```python
# Breaking down the above formula

def rmse(y, y_pred):
    error = y - y_pred # Prediction - the actual value
    se = error ** 2 # square the error value
    mse = se.mean() # get the mean value of all the error values together
    return np.sqrt(mse) # get the square root of that mean
```

```python
# Reimplement the train linear regression algorithm with regularisation

def train_linear_regression_reg(X, y, r=0.0001):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])

    XTX = X.T.dot(X)
    XTX = XTX + r * np.eye(XTX.shape[0])
    
    XTX_inv = np.linalg.inv(XTX)
    w_full = XTX_inv.dot(X.T).dot(y)

    return w_full[0], w_full[1:]
```

```python
# Rerun the validation data set with new features
X_train = prepare_X(df_train)
w0, w = train_linear_regression_reg(X_train, y_train, r=0.1)

X_val = prepare_X(df_val)
y_pred = w0 + X_val.dot(w)

rmse(y_val, y_pred)

# Improved calculation by 0.05 already. Need to find the best value of r though, next lesson
```

### 2.14 - Tuning the Model

Finding the best regularization parameter for the model, r affects the quality of the mdoel so need to find the best one.

```python
for r in [0.0, 0.000001, 0.0001, 0.001, 0.1, 1, 10]:

    # Rerun the validation data set with new features
    X_train = prepare_X(df_train)
    w0, w = train_linear_regression_reg(X_train, y_train, r=r)
    
    X_val = prepare_X(df_val)
    y_pred = w0 + X_val.dot(w)
    
    score = rmse(y_val, y_pred)

    print(r, w0, score)
```

Pick largest r available that does not degrade performance
